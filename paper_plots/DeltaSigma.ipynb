# Generate Table

```python
LABELS = {
    'Maglim' : 'Magnitude Limit',
    'PSF' : 'Seeing',
    'Exptime' : 'Exp. time',
    'SkyBright' : 'Sky brightness',
    'StellarDensity' : 'Star density',
    'Extinction' : 'Galactic Extinction',
}
```

```python
def fmt_pval(x):
    # 2 significant figures, preserving trailing zeros like 0.70
    s = f"{x:#.2g}"
    # Optional: avoid ugly integer-looking values like "70."
    if s.endswith("."):
        s = s[:-1]
    return s


STRING = ""

for model_i, model in enumerate(["Maglim", "PSF", "Extinction"]):

    STR = "V3_"

    # First row of each 4-row block.
    STRING += fr"\multirow{{4}}{{*}}{{ {LABELS[model]} }}"
    STRING += r" & \rdelim\{{4}{*}"

    for name_i, name in enumerate(["DECADE", "DES"]):

        # For the second survey in the same model block, the category+brace columns are empty.
        if name_i > 0:
            STRING += r" & "

        # First row for this survey: survey label + brace + lambda.
        STRING += fr" & \multirow{{2}}{{*}}{{ {name} }}"
        STRING += r" & \rdelim\{{2}{*}"
        STRING += r" & $\lambda$"

        A = np.load(
            f"/scratch/midway3/dhayaa/TMP/{STR}{name}_{model}High_lambda.npy",
            allow_pickle=True,
        )[()]
        B = np.load(
            f"/scratch/midway3/dhayaa/TMP/{STR}{name}_{model}Low_lambda.npy",
            allow_pickle=True,
        )[()]

        for zind in range(len(zbin) - 1):
            
            if (name == 'DECADE') & (zind == 2):
                STRING += f"& ---" * 4
                continue
            
            for lind in range(len(lbin) - 1):
                k = f"{lind}_{zind}"
                
                a = A[k]
                b = B[k]

                s = (a["ds_b"] - b["ds_b"]) / np.sqrt(
                    a["ds_err_b"]**2 + b["ds_err_b"]**2
                )
                mask = a["rp"] > 0.1
                chi2 = np.sum(s[mask] ** 2)
                pval = stats.chi2(df=mask.sum()).sf(chi2)

                STRING += f" & {fmt_pval(pval)}"

        STRING += r" \\" + "\n"

        # Second row for this survey: empty category, brace, survey, survey-brace columns.
        STRING += r" & & & & $\lambda_{\rm cyl}$"

        A = np.load(
            f"/scratch/midway3/dhayaa/TMP/{STR}{name}_{model}High_lambda_cyln.npy",
            allow_pickle=True,
        )[()]
        B = np.load(
            f"/scratch/midway3/dhayaa/TMP/{STR}{name}_{model}Low_lambda_cyln.npy",
            allow_pickle=True,
        )[()]

        for zind in range(len(zbin) - 1):
            
            if (name == 'DECADE') & (zind == 2):
                STRING += f"& ---" * 4
                continue
                
            for lind in range(len(lbin) - 1):
                k = f"{lind}_{zind}"

                a = A[k]
                b = B[k]

                s = (a["ds_b"] - b["ds_b"]) / np.sqrt(
                    a["ds_err_b"]**2 + b["ds_err_b"]**2
                )
                mask = a["rp"] > 0.1
                chi2 = np.sum(s[mask] ** 2)
                pval = stats.chi2(df=mask.sum()).sf(chi2)

                STRING += f" & {fmt_pval(pval)}"

        if name == "DES":
            STRING += r" \\[10pt]" + "\n"
        else:
            STRING += r" \\" + "\n"

print(STRING)
```
